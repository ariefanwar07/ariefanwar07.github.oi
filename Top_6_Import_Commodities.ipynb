```python
import pandas as pd

# Load the uploaded Excel file to examine its structure
file_path = '/content/Indonesia Import.xlsx'
data = pd.ExcelFile(file_path)

# Display sheet names to understand the structure of the file
data.sheet_names

# Load the data from the first sheet to inspect its contents
df = data.parse('Sheet1')

# Display the first few rows to understand the structure of the data
df.head()

# Grouping by HS Code to calculate the total import values for each commodity
top_imports = df.groupby('HS Code')['Import Values'].sum().sort_values(ascending=False).head(6)

# Extracting the result as a DataFrame for better presentation
top_imports_df = top_imports.reset_index()
top_imports_df.rename(columns={'Import Values': 'Total Import Value'}, inplace=True)

# Extracting data for the top 6 commodities across all years
top_6_import_codes = top_imports_df['HS Code'].tolist()
top_6_import_data = df[df['HS Code'].isin(top_6_import_codes)]

# Summarizing yearly import values for the top 6 commodities
top_6_import_yearly = top_6_import_data.pivot_table(index='Year', columns='HS Description', values='Import Values', aggfunc='sum')

# Filling missing values with 0 for better analysis
top_6_import_yearly = top_6_import_yearly.fillna(0)

# Converting the data to long format for better analysis or visualization
top_6_import_long = top_6_import_yearly.reset_index().melt(id_vars='Year', var_name='HS Description', value_name='Import Value')
```

```python
top_6_import_long
```

```python
output_file_path = 'Top_6_Import_Commodities.xlsx'
top_6_import_long.to_excel(output_file_path, index=False)

print(f"Data has been saved to {output_file_path}")
```

```python
# Step 1: Summarize yearly import values for the top 6 commodities
total_import_yearly = df.pivot_table(index='Year', columns='HS Description', values='Import Values', aggfunc='sum')

# Step 2: Calculate the total import per year
total_import_per_year = total_import_yearly.sum(axis=1)

# Step 3: Calculate the percentage import values
top_6_import_percentage = (top_6_import_yearly.div(total_import_per_year, axis=0)).fillna(0)

# Step 4: Convert the data to long format for better analysis
top_6_import_percentage_long = top_6_import_percentage.reset_index().melt(id_vars='Year', var_name='HS Description', value_name='Import Percentage')
top_6_import_percentage_long
```

```python
output_file_path = 'Top_6_Import_Commodities_Percentage.xlsx'
top_6_import_percentage_long.to_excel(output_file_path, index=False)

print(f"Data has been saved to {output_file_path}")
```

```python
top_6_import_percentage_long['Year'] = pd.to_datetime(top_6_import_percentage_long['Year'], format='%Y')
```
